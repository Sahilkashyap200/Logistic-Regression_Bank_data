```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
data=pd.read_csv("\\Users\\piyus\\Documents\\DataSets\\assignment\\bank.csv")
data.head()
```

```python
data_=data.copy()
```

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
data_.iloc[:,-1]=label.fit_transform(data_.iloc[:,-1])
```

```python
data_
```

```python
data_=data_.drop('contact',axis=1)
data_=data_.drop('day',axis=1)
data_=data_.drop('month',axis=1)
data_=data_.drop('duration',axis=1)
data_=data_.drop('campaign',axis=1)
```

```python
data_.shape
```

```python
data_=pd.get_dummies(data_.iloc[:,0:12])
```

```python
data_.head()
```

```python
def norm_func(i):
 x=(i-i.min())/(i.max()-i.min())
 return (x)
```

```python
dataframe=norm_func(data_)
```

```python
#split data into x and Y
X=dataframe.drop('deposit',axis=1)
Y=dataframe['deposit']
```

```python
#split data into training and testing set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=77)

x_train.shape,x_test.shape
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
